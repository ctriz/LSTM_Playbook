```python
from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, LSTM
from keras.layers import GlobalMaxPooling1D
from keras.models import Model
from keras.layers.embeddings import Embedding
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.layers import Input
from keras.layers.merge import Concatenate
from keras.layers import Bidirectional

import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
```

```python
#Many-to-One 
#Each time step has one feature
```

```python
nums = 25

X1 = list()
X2 = list()
X = list()
Y = list()

X1 = [(x+1)*2 for x in range(25)]
X2 = [(x+1)*3 for x in range(25)]
Y = [x1*x2 for x1,x2 in zip(X1,X2)]

print(X1)
print(X2)
print(Y)
```

```python
#The input is a combination of two lists X1 and X2. T
#The output is the product of X1*X2
```

```python
#Create the input
X = np.column_stack((X1, X2))
print(X)
```

```python
# The input set consists of two features per input
# Now convert the input into 3-dimensional shape. 
# Input has 25 samples, where each sample consist of 1 time-step and each time-step consists of 2 features. 
```

```python
# reshape the input and output
X = array(X).reshape(25, 1, 2)
Y = array(Y).reshape(25, 1, 1)
```

```python
model = Sequential()
model.add(LSTM(100, activation='relu',return_sequences=True, input_shape=(1, 2)))
model.add(LSTM(25, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
print(model.summary())
```

```python
model.fit(X, Y, epochs=100,validation_split=0.2, batch_size=5, verbose =1) 
```

```python
# validation_split: Float between 0 and 1. 
# Fraction of the training data to be used as validation data. 
# The model will set apart this fraction of the training data, will not train on it.
# Will evaluate the loss and any model metrics on this data at the end of each epoch. 
# The validation data is selected from the last samples in the x and y data provided, before shuffling.
```

```python
# Our test data point will have two features i.e. (55,80) the actual output should be 55 x 80 = 4400.
```

```python
test_input = array([55,80])
test_input = test_input.reshape((1, 1, 2)) # reshaped into 1 sample, 1 time step, 2 feature
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

```python
# Many-to-One Sequence Problems with a Single Feature
```

```python
X = np.array([x+1 for x in range(45)])
print(X)
```

```python
#15 samples, 3 time-steps, and 1 feature.
X = X.reshape(15,3,1) 
print(X)
```

```python
# each element in the output will be equal to
# the sum of the values in the time-steps in the corresponding input sample
```

```python
Y = list()
for x in X:
    Y.append(x.sum())

Y = np.array(Y)
print(Y)
```

```python
model = Sequential()
model.add(LSTM(100, activation='relu',return_sequences=True, input_shape=(3, 1))) #3 time step, 1 feature
model.add(LSTM(25, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
print(model.summary())
```

```python
history = model.fit(X, Y, epochs=100, validation_split=0.2, verbose=1)
```

```python
test_input = array([50,51,52])
test_input = test_input.reshape((1, 3, 1)) #sample size=1, time step =3, feature =1
test_output = model.predict(test_input, verbose=0)
print(test_output)
```
